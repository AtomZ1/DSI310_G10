## Getting data 

```python
import bs4 
import requests
import pandas as pd
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
import json

data = urlopen('https://www.harpersbazaar.com/search/?q=jewelry+trends+2023')
s = soup(data.read(),features='html.parser')

title_list = []
details_list = []

headline = s.find_all('div',attrs= {'class' : 'simple-item-title item-title'})
for each in headline:
    title_list.append(str(each.get_text()))
    

paragraph = s.find_all('div',attrs= {'class' : 'simple-item-dek item-dek'})
for each in paragraph:
    details_list.append(str(each.get_text().replace('\n','')))
    
table = pd.DataFrame([title_list,details_list]).transpose()
table.columns = ['title','detail']
```

```python
table = table.to_dict('records')
print (f"page:{len(table)}")
```
